```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import os


path = "../../loc2vec/tiles/tiles-dataset-v2.npy"

# images = np.load(path, allow_pickle=True)
images = np.lib.format.open_memmap(path, mode='r', dtype=np.uint8)
print(f"Loaded {len(images)} images from {path}")

sample_size = 10 * 10
sample_indices = np.random.choice(len(images), sample_size, replace=False)


from torchvision.utils import make_grid
import torch

def plot_images(images, ncols=10):
    nrows = (len(images) + ncols - 1) // ncols
    grid = make_grid(images, nrow=ncols, padding=2, normalize=True)
    plt.figure(figsize=(15, 15))
    plt.imshow(grid.permute(1, 2, 0).numpy())
    plt.axis('off')
    plt.show()


sample_images = images[sample_indices].astype(np.float32)
plot_images(torch.from_numpy(sample_images))

```
